```python
def vect(x):
  return np.array(x)
```

```python
# Pendefinisian Vektor

A = [17,22,19]
B = [10,20,11]
C = [5,12,9]

vect_A = vect(A)
vect_B = vect(B)
vect_C = vect(C)

print('Vektor A adalah \n', vect_A, '\n')
print('Vektor B adalah \n', vect_B, '\n')
print('Vektor C adalah \n', vect_C, '\n')
```

```python
# A + B
print(vect_A, '+', vect_B , '=', vect_A + vect_B) 

# B - C
print(vect_B, '-', vect_C, '=', vect_B - vect_C) 

# A dot C
print(vect_A, '.', vect_C, '=', vect_A @ vect_C)

# A x B
print(vect_A, 'x', vect_B, '=', vect_A * vect_B) 

# Norm A atau ||A||
print('Hasil dari norm A adalah =', np.linalg.norm(vect_A)) 
```

```python
# Sudut antara Vektor A dan Vektor B

theta = vect_A @ vect_B / (np.linalg.norm(vect_A) * np.linalg.norm(vect_B))

print('Hasil theta yang didapat yaitu:', theta)
```

```python
# Mencari nilai sudut dengan mencari nilai radian dahulu, kemudian mencari nilai sudutnya.

radian = np.arccos(theta)
degree = np.degrees(radian)

print('Nilai sudut antar Vektor A dan Vektor B adalah', degree)
```
