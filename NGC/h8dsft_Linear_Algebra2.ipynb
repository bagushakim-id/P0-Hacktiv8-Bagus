```python
A = [[23,50,19],
     [7,12,109],
     [57,67,98]]

B = [17,22,19]

matrix_A = np.array(A)
matrix_B = np.array(B)

print('Ini adalah matrix A = \n', matrix_A, '\n')
print('Ini adalah matrix B = \n', matrix_B, '\n')
print('Berikut hasil dari perkalian matrix A dengan Matrix B:')
print(matrix_A, 'x', matrix_B, '=', matrix_A @ matrix_B)
```

## Soal 3

image.png

```python
A = [[23,50,19],
     [7,12,109],
     [57,67,98]]

matrix_A = np.array(A)
```

```python
# Transpose

print('Matrix A:')
print(matrix_A)
print('')
print('Matrix A transpose:')
print(matrix_A.transpose())
```

```python
# Determinan
det = np.linalg.det(matrix_A)

print('Determinan matrix A adalah:', det)
```

```python
# Inverse
Inv_matrix_A = np.linalg.inv(matrix_A)

print('Inverse dari matrix A adalah:')
print(Inv_matrix_A)
```
